```python
import cv2
import cvzone
import numpy as np
import tkinter as tk
from tkinter import filedialog
from cvzone.HandTrackingModule import HandDetector


cap = cv2.VideoCapture(1)
cap.set(3, 2560)
cap.set(4, 1440)
detector = HandDetector(detectionCon=1)

root=tk.Tk()
root.withdraw()


colorR = (255, 0, 255)
cx, cy, w, h = 100, 100, 200, 200


class btns():
    def __init__(self, posCenter, value, size=[200, 100]):
        self.posCenter = posCenter
        self.size = size
        self.value = value
    def select(self,cursor):
        cx, cy = self.posCenter
        w, h = self.size
        value=self.value

        # If the index finger tip is in the rectangle region
        if cx - w // 2 < cursor[0]< cx + w // 2 and \
                cy - h // 2 < cursor[1]< cy + h // 2:
            if value=="select file":
                file_path=filedialog.askopenfilename()
                print(file_path)
            if value=="select files":
                file_path=filedialog.askopenfilenames()
                print(file_path)
            if value=="volume":
                print("ok")

        
btnvalues=["select file","select files","volume"]
btnList = []
i=0
for x in btnvalues:
    btnList.append(btns([120 ,i * 150 +  80],x))
    i+=1


while True:
    success, img = cap.read()
    img1 = cv2.flip(img, 1)
    lmList = detector.findHands(img1,draw=False)
    if lmList:
        if len(lmList)==1:
            hand=lmList[0]
            _,cursor= detector.findDistance(hand["lmList"][8], hand["lmList"][12])
            if   detector.fingersUp(hand)==[0,1,1,0,0] :
                for btn in btnList:
                    btn.select(cursor[4:])
                    
    
    
    
    
    
    
    
    
    imgNew = np.zeros_like(img1, np.uint8)
    for btn in btnList:
        cx, cy = btn.posCenter
        w, h = btn.size
        cv2.rectangle(imgNew, (cx - w // 2, cy - h // 2),
                      (cx + w // 2, cy + h // 2), colorR, cv2.FILLED)
        cv2.putText(imgNew,btn.value,(cx - w // 2, cy ),cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0))
        cvzone.cornerRect(imgNew, (cx - w // 2, cy - h // 2, w, h), 20, rt=0)

    out = img1.copy()
    alpha = 0.5
    mask = imgNew.astype(bool)
    out[mask] = cv2.addWeighted(img1, alpha, imgNew, 1 - alpha, 0)[mask]
    
    
    cv2.imshow("Image", out)
    if cv2.waitKey(27)==27:
        cap.release()
        cv2.destroyAllWindows()
        
        break


```
